```python
import numpy as np
import random as rm
import matplotlib.pyplot as plt
from IPython.display import Image
import timeit
```

```python
states = ["Death", "Life", "Start"]
transitionName = [["NN", "NR", "ND"], ["RN", "RR", "RD"], ["DN", "DR", "DD"]]
transitionMatrix = [[0, .9, .1], [1, 0, 0], [1, 0, 0]]
```

```python
def forecast_population(hours):
    start = timeit.default_timer()
    initialState = "Start"
    i = 0
    n = 1
    while i != hours:
        if initialState == "Start":
            change = np.random.choice(transitionName[0], replace=True, p=transitionMatrix[0])
            if change == "NR":
                initialState = "Life"
                n += 1
                graphArray.append(n)
            else:
                initialState = "Death"
                n -= 1
                graphArray.append(n)
        elif initialState == "Life" or initialState == "Death":
            initialState = "Start"
        i += 1
    print("Final population: " + str(n))
    stop = timeit.default_timer()
    print('Compile time: ', stop - start)  
```

```python
graphArray = [0]
forecast_population(10000)
```

```python
plt.xlabel('Time')
plt.ylabel('Population')
plt.plot(graphArray)
plt.show()
```
