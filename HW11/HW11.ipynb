### Variables:
n = Number of diodes per test group<br>
C = Testing cost for one group<br>
A = Average testing cost<br>

### Assumptions:
if n=1, A =4,<br>
otherwise (n>1), we have C = 3+n<br>
if all diodes are good, and C = (3+n)+4n <br>
if the group indicates a failure.<br>
A = (Average value of C)/n

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
import sympy as sp
```

## Part 1:

```python
#Just to display function below
n = sp.symbols('n')
A_f = 3/n+5-4*.998**n

#Defining function to graph
n = np.arange(1, 255, .01)
def A(x):
    return (3/x)+5-4*(.998)**x
```

```python
plt.xlabel('number of diodes per test group')
plt.ylabel('Average testing cost')
plt.plot(n, A(n))
plt.show()
```
