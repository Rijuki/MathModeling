# Homework problem 4

### Variables:

    S = Number of Phones sold
    TC = Total Cost (dollars)
    Y = Additional dollars spent of Youtube Advertising (dollars)
    d = discount from MSRP of 850 dollars
    p = profit (dollars)
    Ld = lambda
    
### Assumptions:

```python
import sympy as sp
import numpy as np

S, TC, Y, d, p, Ld = sp.symbols('S TC Y d p Ld')

S = (((15000)*(1+.004*d)) + (2*Y)/5) - 500
TC = ((600*S) + (Y + 4000))
P = ((850 - d)*S) - TC
```

```python
P
```

```python
dPdY = Ld * sp.diff(P, Y)
dPdd = Ld * sp.diff(P, d)
```

```python
dPdY
```

```python
dPdd
```

```python
%matplotlib inline
from sympy.plotting import plot3d

plot3d(P, (Y, 0, 10000), (d, 0, 240),show=True)
```

```python
Soln = sp.nonlinsolve([dPdY, dPdd], (Ld, d, Y))
```
