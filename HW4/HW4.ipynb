### Assumptions:

```python
import sympy as sp
import numpy as np

S, TC, Y, d, p, Ld = sp.symbols('S TC Y d p Ld')

Y1 = Y/2000
S = ((500*sp.log(Y1, 2))+((15000)*(1+.004*d)) + Y/5)
TC = ((600*S) + (Y + 4000))
P = ((850 - d)*S) - TC
```

```python
dPdY = Ld * sp.diff(P, Y)
dPdd = Ld * sp.diff(P, d)
```
