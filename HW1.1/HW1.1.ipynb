**Question**: At what rebate amount will maximize revenue?

**Modeling Approach**: Calculus optimization techniques.

Formulated model below:

```python
#imported libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative

#Defines the model and its range
x1 = np.arange(-7, 7, .01)
def Y(x):
    return (140 - 20 * x)*(1 + .16 * x)

#Plots the defined Y function 
plt.xlabel('# of rebates')
plt.ylabel('Profit')
plt.plot(x1, Y(x1))
plt.show()
```

Solve the Model:<br>
Graph is extended into negative numbers to better display model.

```python
#Defines a function that takes the derivative and plots it
x2 = np.arange(0, 10, 1)
def yprime(x):
    return derivative(Y, x)

plt.plot(yprime(x2))

for x2 in np.arange(0, 10, .001):
    if yprime(x2) == 0:
        print('The max revenue is at {:.03f}'.format(x2))

plt.xlabel('# of rebates')
plt.ylabel('Profit')
plt.show()        
```
