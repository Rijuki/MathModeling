```python
#imported libraries
import numpy as np
import sympy as sp
from scipy.misc import derivative

H, S = sp.symbols('H S')

r1 = .1
r2 = .25
a1 = .1/10000
a2 = .25/6000
b1 = .1/(2*10000)
b2 = .25/(2*6000)

dHdt = r1*H-a1*H**2-b1*S*H
dSdt = r2*S-a2*S**2-b2*S*H

T = dHdt + dSdt
```

```python
dTdH = sp.diff(T, H)
dTdS = sp.diff(T, S)
```
